```python
import os
import numpy as np
from scipy.io import savemat

def convert_npy_to_mat(source_dir, target_dir):
    """
    Convert all .npy files in source_dir to .mat files in target_dir,
    maintaining the directory structure.
    """
    for root, dirs, files in os.walk(source_dir):
        # Construct the target directory path
        rel_path = os.path.relpath(root, source_dir)
        target_path = os.path.join(target_dir, rel_path)
        
        # Create the target directory if it does not exist
        os.makedirs(target_path, exist_ok=True)
        
        for file in files:
            if file.endswith('.npy'):
                # Construct full file paths
                npy_path = os.path.join(root, file)
                mat_path = os.path.join(target_path, file.replace('.npy', '.mat'))
                
                # Load the .npy file and convert to .mat
                data = np.load(npy_path)
                savemat(mat_path, {'data': data})
                
                print(f"Converted {npy_path} to {mat_path}")

# Example usage
source_dir = '/Users/alexa/Documents/MATLAB/U-SPDNet/UAV_CONV_orig'  # Your source directory
target_dir = '/Users/alexa/Documents/MATLAB/U-SPDNet/UAV_CONV_MAT_orig'  # Your target directory for .mat files

convert_npy_to_mat(source_dir, target_dir)
```

```python
import os
import numpy as np
from scipy.io import savemat

# Base directory for UAV_COV data
base_dir = '/Users/alexa/Documents/MATLAB/U-SPDNet/data/afew/UAV_CONV'

# Labels and their corresponding numeric values
label_map = {'1': 1, '2': 2, '3': 3}

# Initialize the spd structure
spd = {'name': [], 'id': [], 'set': [], 'label': []}

# Function to process and add data to the spd structure
def process_data(sub_dir, set_val):
    for label_dir in os.listdir(sub_dir):
        label_path = os.path.join(sub_dir, label_dir)
        if os.path.isdir(label_path):
            for file in os.listdir(label_path):
                if file.endswith('.mat'):
                    # Construct file path
                    filepath = os.path.join(label_path, file)
                    # Add the entry to the spd structure
                    relative_path = os.path.relpath(filepath, base_dir)
                    spd['name'].append(relative_path)
                    spd['id'].append(len(spd['id']) + 1)
                    spd['set'].append(set_val)
                    spd['label'].append(label_map[label_dir])

# Process train and validation data
process_data(os.path.join(base_dir, 'train'), 1)  # Training data
process_data(os.path.join(base_dir, 'val'), 2)    # Validation data

# ... [previous code]

# Convert lists to arrays for MATLAB compatibility
for key in ['name']:
    spd[key] = np.array(spd[key], dtype=object)

for key in ['id', 'set', 'label']:
    spd[key] = np.array(spd[key], dtype=np.double)

# The SpdDir structure containing the file path
SpdDir = np.array(['./data/afew/UAV_CONV'], dtype=object)

# Create the final structure to be saved
spd_train = {'spd': spd, 'SpdDir': SpdDir}

# Save the spd_train structure to a .mat file
savemat('sample_for_SPDNet_UAV.mat', {'spd_train': spd_train})
```
